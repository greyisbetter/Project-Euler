```python
# prime number to n

def prime_number(n):
    number_list = list()
    for x in range(2, n):
        temp = x
        for y in range(2, x):
            if (x % y) == 0:
                temp = None
        if temp != None:
            number_list.append(temp)
    return number_list
```

```python
prime_number(10)
```

```python
# factorial of n

def factorial(n):
    fact = 1
    if n < 0:
        return "less than 0"
    elif n == 0:
        return 0
    else:
        for i in range(1, n+1):
            fact *= i
    return fact
```

```python
# function to check prime number

def isPrime(n):
    flag = True
    if n < 2:
        flag = False
    else:
        for i in range(2, n):
            if (n % i) == 0:
                flag = False
    return flag
```

```python
# divide number from bet of its factors
def divideBet(n):
    # don't know why
    # may be the first quotient of n : n / 1 == n
    quotient = n
    # loop for 1->n
    for i in range(1, n+1):
        # x = (quot, rem)
        x = divmod(n, i)
        # if completly divisible
        if x[1] == 0:
            # i is factor of n
            divisor = i
            # check for divisor == quotient situation eg. 4 = 2x2
            if divisor == x[0]:
                quotient = x[0]
                break
            
            # if not check for persent divisor == prev. quotient eg. 48 = 8x6
            if divisor == quotient:
                divisor = x[0]
                break
            
            # if both not then continue with new quotient value
            quotient = x[0]
            
    return (divisor, quotient)
```

```python
# prime factors will be leaves of binary tree were nodes are division of numbers

# object Node
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
# function to create a binary tree using prime factor algorithms

def insertion(n, node):
    # first get a right and left value for nodes of n
    left, right = divideBet(n)
    
    # check if left value is equal to 1 i.e., the node is prime number
    if left == 1: return
    
    # insert left and right node of n node
    node.left = Node(left)
    node.right = Node(right)
    
    # now recursion for left and right node
    insertion(left, node.left)
    insertion(right, node.right)
```

```python
# get leaves of binary tree which are prime factors

def leaves(node, leaf):
    # if the node is last node which have not left and right node
    if (not node.left) and (not node.right):
        leaf.append(node.data)
        return
    
    # recurse for left and right node
    leaves(node.left, leaf)
    leaves(node.right, leaf)    
```
