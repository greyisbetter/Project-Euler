```python
def checkPalindromicNumber(num):
    number = str(num)
    reversedNumber = number[::-1]
    
    return number == reversedNumber
```

```python
checkPalindromicNumber(9509)
```

```python
xArray = []
yArray = []
productArray = []
for x in reversed(range(1000)):
    for y in reversed(range(1000)):
        product = x * y
        if checkPalindromicNumber(product):
            xArray.append(x)
            yArray.append(y)            
            productArray.append(product)     
            
ans = max(productArray)
place = productArray.index(ans)
operands = (xArray[place], yArray[place])
print('x'.join(list(map(str, operands))), '=', ans)
```
