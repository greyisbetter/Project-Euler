```python
# fibonacci sequence of first n term:
def fib_nterm(n_term):
    a = 1
    b = 2
    fib_list = []
    while len(fib_list) < n_term:
        fib_list.append(a)
        fib_list.append(b)
        a += b
        b += a
    return b
```

```python
def fib_sum(a, b, n):
    sum = 0
    while b < n:
        if a % 2 == 0:
            sum += a
        if b % 2 == 0:
            sum += b
        a += b
        b += a
    return sum
```

```python
# test for first 10 term
fib_sum(1, 2, 89)
```

```python
# now test for 4M

fib_sum(1, 2, 4000000)
```

```python
fib_sum(1, 2, fib_nterm(10))
```
